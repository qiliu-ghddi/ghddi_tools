```python
from ete3 import Tree
from IPython.display import Image

# 创建一个示例树
t = Tree('((a,b),c);')

# 渲染为PNG图片并显示
t.render("tree.png")
Image(filename='tree.png')
```

```python
import ete3
t = ete3.Tree()
t.populate(20, ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'], random_branches=True, reuse_names=True)
print(t)
```

```python
t.write(outfile='demo_tree2.txt')
```

```python
# t.render("demo_tree2.png", w=183, units="mm")
```

```python
from ete3 import Tree
with open('demo_tree2.txt', 'r') as file:
    s = file.read() # read from file
    t = Tree(s) # read Newick format 8
    print(t) # print tree as txt
#     t.render('demo_tree2.png') # export png image
```

## Phylo

```python
from io import StringIO
from Bio import Phylo

treedata = "(A, (B, C), (D, E))"
treedata = "((a,b),c);" 
handle = StringIO(treedata)
tree = Phylo.read(handle, "newick")

print(tree)
```

```python
Phylo.draw(tree)
```

```python
import pylab
Phylo.draw_graphviz(tree)
# pylab.show()
```

```python
Phylo.draw_ascii(tree)
```

## Plotly

```python
import plotly.figure_factory as ff

import numpy as np

X = np.random.rand(15, 10) # 15 samples, with 10 dimensions each
fig = ff.create_dendrogram(X, color_threshold=1.5)
fig.update_layout(width=800, height=500)
fig.show()
```
